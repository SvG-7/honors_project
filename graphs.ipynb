```python
import numpy as np
import matplotlib.pyplot as plt

points = np.loadtxt('points_10.txt')

boundary_points = np.loadtxt('hull_output.txt')

plt.figure(figsize=(10, 6))
plt.scatter(points[:, 0], points[:, 1], s=10, label='Points')

boundary_points = np.vstack([boundary_points, boundary_points[0]])  
plt.plot(boundary_points[:, 0], boundary_points[:, 1], color='red', label='Polygon')

plt.title('Points and Polygon')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
